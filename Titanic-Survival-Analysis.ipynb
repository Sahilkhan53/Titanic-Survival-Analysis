## 6) Implement ideal table properties

#### Column Level

```python
df.rename(columns={'Name': 'Passenger_Name'}, inplace=True)
df.drop(['Ticket', 'Cabin'], axis=1, inplace=True, errors='ignore')
```

#### Row Level

```python
df.drop_duplicates(inplace=True)
```

#### Value Level

```python
df['Sex'] = df['Sex'].astype('category')
```

####  Confirm

```python
df.isnull().sum()
```

## 7) Fill Null Values

#### Numerical

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

#### Categorical

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

## 8.1) Numerical vs Numerical Analysis

```python
sns.scatterplot(x='Age', y='Fare', data=df)
plt.title('Age vs Fare')
plt.show()
```

## 8.2) Categorical vs Categorical Analysis

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival Count by Gender')
plt.show()
```

## 8.3) Numerical vs Categorical Analysis

```python
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Age Distribution by Survival')
plt.show()
```

## 9) Visualize the date with appropriate charts

```python
sns.countplot(x='Survived', data=df)
plt.title('Overall Survival Count')
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival by Passenger Class')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Gender-wise passenger count
gender_counts = df['Sex'].value_counts()

# Pie chart
plt.figure(figsize=(6,6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140, colors=['lightblue', 'lightcoral'])
plt.title('Gender Distribution of Passengers')
plt.axis('equal')  # Circle shape
plt.show()
```

```python
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

### Conclusion
- Females had a higher survival rate than males.
- Passengers in 1st class were more likely to survive.
- Children had better survival chances.
- Higher fare passengers generally survived more.
- Passengers who embarked at port 'C' had slightly better survival rates.
